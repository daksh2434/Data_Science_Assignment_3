# **Project Summary -**

The Global Terrorism Dataset provides a comprehensive record of terrorist incidents that occurred worldwide between 1970 and 2017. This dataset contains detailed information on over 180,000 attacks, including attributes such as date, location, attack type, target type, weapon used, terrorist group involved, and the number of casualties. The dataset serves as a valuable resource for understanding patterns, trends, and the evolving nature of terrorism across different regions and time periods.

The primary objective of this project is to analyze the dataset using exploratory data analysis (EDA) techniques to uncover meaningful insights related to global terrorism. The analysis focuses on identifying trends over time, understanding the geographic distribution of attacks, examining the most common attack methods, and evaluating the impact in terms of casualties and damage.

The project begins with data loading and preprocessing. Missing values in important numerical columns such as the number of people killed and wounded were handled by replacing them with zero. A new feature, “casualties,” was created by combining the number of killed and wounded to better measure the impact of each attack. The dataset was also cleaned and formatted to ensure consistency and accuracy for analysis.

Various visualization techniques were used to explore the data, including bar charts, pie charts, count plots, box plots, histograms, and heatmaps. These visualizations helped in understanding both univariate and multivariate relationships within the dataset. For example, line and bar charts were used to analyze trends in the number of attacks over the years, while pie charts and count plots helped examine the distribution of attacks across regions and attack types. Box plots were used to study the distribution and spread of casualties, highlighting the presence of extreme outliers. Heatmaps and pair plots were utilized to analyze correlations and relationships between numerical variables such as kills, wounds, and total casualties.

The analysis revealed several important insights. There has been a significant increase in the number of terrorist attacks over time, particularly after 2010, with a peak around 2013–2014. The data also shows that terrorism is highly concentrated in specific regions, especially the Middle East & North Africa and South Asia. Countries like Iraq, Pakistan, and Afghanistan have experienced the highest number of attacks. Among attack types, bombing and explosions are the most common methods used by terrorist groups. Although most attacks result in relatively low casualties, a small number of incidents cause extremely high damage, indicating a highly skewed distribution.

Further analysis showed that a limited number of terrorist groups are responsible for a large proportion of attacks, and their activities are often region-specific. Strong correlations were observed between the number of people killed, wounded, and total casualties, confirming that these variables are directly related and critical for assessing the severity of attacks.

Overall, this project demonstrates how data analysis can be used to gain valuable insights into global terrorism. The findings can help governments, organizations, and policymakers make informed decisions regarding security planning, resource allocation, and risk management. By understanding the patterns and factors associated with terrorist activities, more effective preventive measures and strategies can be developed to reduce the impact of such incidents in the future.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df['nkill'] = df['nkill'].fillna(0)
df['nwound'] = df['nwound'].fillna(0)
df['casualities'] = df['nkill'] + df['nwound']
df['iyear'] = df['iyear'].astype(int)

attacks_per_year = df['iyear'].value_counts().sort_index()

top_groups = df['gname'].value_counts().head(5).index
filtered = df[df['gname'].isin(top_groups)]

region_counts = df['region_txt'].value_counts()
````

### What all manipulations have you done and insights you found?

Manipulations Done :
* Missing values in nkill and nwound were replaced with 0 to handle incomplete data.
* A new feature "casualities" was created by combining number of killed and wounded.
* The year column (iyear) was converted into integer format for proper analysis.
* Year-wise attack counts were calculated to analyze trends over time.
* Top 5 terrorist groups were selected based on frequency of attacks.
* Data was filtered to focus only on these top groups for deeper analysis.
* Region-wise attack counts were calculated to understand geographic distribution.


---

Insights Found :
* A small number of terrorist groups are responsible for a large proportion of total attacks.
* Casualties provide a better measure of attack impact by combining kills and injuries.
* Most attacks result in low casualties, but a few incidents cause extremely high damage.
* Terrorism is highly concentrated in specific regions rather than evenly spread globally.
* Certain regions consistently experience higher attack frequencies.
* Focusing on top groups helps identify dominant patterns and major contributors to terrorism.
